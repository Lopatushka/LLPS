```python
import pandas as pd
from pathlib import Path
```

```python
dir = "/mnt/c/users/Elena/Desktop/Data_processing/280226_U2OS_live_FRAP2/results"
folder = Path(dir)
```

```python
dfs = []  # list of DataFrames

for file in folder.glob("*.csv"):
    df = pd.read_csv(file)
    dfs.append(df)
    print(file, df.shape[0])

print(f"Loaded {len(dfs)} files.")
```

```python
N = 43
trimmed_dfs = {}
output_folder = "/mnt/c/users/Elena/Desktop/Data_processing/280226_U2OS_live_FRAP2/results/trimmed"
for file in folder.glob("*.csv"):
    df = pd.read_csv(file)
    df_trimmed = df.iloc[:N].copy()
    trimmed_dfs[file.stem] = df_trimmed

    # save file
    df_trimmed.to_csv(output_folder +  f"/{file.stem}_trimmed.csv", index=False)
```
